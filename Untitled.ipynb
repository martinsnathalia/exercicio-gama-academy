```python
# Importando a biblioteca
import pandas as pd
```

```python

# Leitura dos dados
df = pd.read_csv('pokemon.csv')
print(df)
```

```python
df
```

```python
# Convertendo o tipo de dado
df['Total'] = pd.to_numeric(df['Total'])
df['HP'] = pd.to_numeric(df['HP'])
df['Attack'] = pd.to_numeric(df['Attack'])
df['Defense'] = pd.to_numeric(df['Defense'])
df['Sp. Atk'] = pd.to_numeric(df['Sp. Atk'])
df['Sp. Def'] = pd.to_numeric(df['Sp. Def'])
df['Speed'] = pd.to_numeric(df['Speed'])
df['Generation'] = pd.to_numeric(df['Generation'])
```

```python
# Preenchendo valores nulos para 0
df['Type 2'] = df['Type 2'].fillna(0)
```
